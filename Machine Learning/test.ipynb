```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords  # Import for stop words (optional)


def recommend_skills(data, username, k=5):
  """
  Recommends programming skills for a user based on their skills and similarity to other users.

  Args:
      data: Pandas DataFrame containing user data with username and skill columns.
      username: Username of the user for whom recommendations are needed.
      k: Number of recommendations to return (default: 5).

  Returns:
      A list of recommended skills (strings).
  """

  # Select relevant columns (username and skill columns)
  relevant_skill_columns = data.columns[1:]  # Exclude username

  # Filter rows with empty skills (optional)
  data_filtered = data[data["skills_string"].notna()]

  # Preprocess skill data (create comma-separated string)
  def filter_and_join(x):
      return ", ".join(str(val) for val in x.dropna() if isinstance(val, str))

  data_filtered["skills_string"] = data_filtered[relevant_skill_columns].apply(filter_and_join, axis=1)

  # One-hot encode skills (consider stop words handling)
  # Option 1: Disable stop word removal
  # vectorizer = CountVectorizer(stop_words=None)

  # Option 2: Customize stop words (download NLTK stopwords first)
  custom_stop_words = stopwords.words('english')
  # Add specific LeetCode terms to stop words if needed (e.g., 'Java', 'Python')
  vectorizer = CountVectorizer(stop_words=custom_stop_words)

  try:
      skill_matrix = vectorizer.fit_transform(data_filtered["skills_string"])
  except ValueError as e:
      if "empty vocabulary" in str(e):
          print("Error: Empty vocabulary encountered. Consider adjusting stop words or data preprocessing.")
          return []  # Handle empty vocabulary gracefully (no recommendations)
      else:
          raise e  # Re-raise other ValueErrors

  # Calculate user similarity
  similarity_matrix = cosine_similarity(skill_matrix)

  # Find user data
  user_data = data_filtered[data_filtered["username"] == username]

  # Handle cases with single or multiple user rows
  if len(user_data) == 1:
      user_index = user_data.index.values[0]
      user_skills = user_data["skills_string"].values[0]  # Assuming skills_string
  else:
      # Combine skills from all rows (assuming unique skills)
      user_skills = set(user_data[relevant_skill_columns].values.flatten())

  # Calculate similar users
  user_index = user_data.index.values[0]
  similar_users = similarity_matrix[user_index].argsort()[-k:][::-1]

  # Extract skills from neighbors
  neighbor_skills = data_filtered.loc[similar_users, relevant_skill_columns].values.tolist()
  neighbor_skills_flat = [item for sublist in neighbor_skills for item in sublist]  # Flatten

  # Filter recommended skills (remove user's skills)
  unique_skills = set(neighbor_skills_flat) - set(user_skills.split(", "))

  # Filter skills within vocabulary range
  valid_skills = unique_skills.intersection(set(range(len(vectorizer.get_feature_names_out()))))

  # Map recommended skill indices to skill names (using valid skills)
  recommended_skill_names = [vectorizer.get_feature_names_out()[idx] for idx in valid_skills]

  return recommended_skill_names


# Load data (replace with your CSV path)
data = pd.read_csv('../Leetcode.csv')
data.head()

# Example usage
username = "sudhanshu_090"
k = 10  # Number of recommendations

recommended_skills = recommend_skills(data.copy(), username, k)
print(f"Recommended skills for {username}: {', '.join(recommended_skills)}")
```
