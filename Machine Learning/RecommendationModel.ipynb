```python
import pandas as pd
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from surprise import accuracy
import requests
import json
from datetime import datetime, timezone, timedelta
```

```python



df = pd.read_csv('../Leetcode.csv')
df.head()
df = df.drop(['total','easy','medium','hard','rating','C++','Java','Python','Python3','C','C#','JavaScript','TypeScript','PHP','Swift','Kotlin','Dart','Go','Ruby','Scala','Rust','Racket','Erlang','Elixir','MySQL','MS SQL Server','Oracle','Pandas','PostgreSQL','ranking','contestAttended','streak','totalActiveDays','OurRating'],axis=1)
df.head()
df.dtypes
```

```python
# Define the rating scale
reader = Reader(rating_scale=(1, 5))


# Reshape the DataFrame into long format
df_long = df.melt(id_vars=['username'], var_name='skill', value_name='number_of_questions')
# df_append =[]

# for i in df_long['number_of_questions']:
       
           
#         if i>15:
#             df_append.append(2)
#         else:
#             df_append.append(1)

# df_long['number_of_questions'] = df_append

# Filter out rows where the rating is 0 or NaN (if applicable)
# df_long = df_long[df_long['rating'] != 0]
            
print(df_long)

# Load the dataset
dataset = Dataset.load_from_df(df_long[['username', 'skill', 'number_of_questions']], reader)
print(dataset)
```

```python





# Load the dataset
# dataset = Dataset.load_from_df(df[['username','Array', 'String',"Hash Table","Math","Dynamic Programming","Sorting","Greedy","Depth-First Search","Binary Search","Database","Breadth-First Search","Tree","Matrix", "Two Pointers", "Bit Manipulation","Binary Tree","Heap (Priority Queue)","Stack","Prefix Sum","Simulation","Graph","Design","Counting","Sliding Window","Backtracking", "Union Find","Enumeration", "Linked List", "Ordered Set", "Monotonic Stack","Trie","Number Theory", "Divide and Conquer", "Recursion", "Bitmask",  "Queue", "Binary Search Tree","Segment Tree", "Memoization", "Hash Function", "Geometry", "Binary Indexed Tree", "Topological Sort",  "String Matching","Combinatorics","Rolling Hash","Shortest Path","Game Theory","Data Stream","Interactive","Brainteaser","Monotonic Queue", "Randomized","Merge Sort", "Iterator", "Concurrency","Doubly-Linked List","Probability and Statistics","Quickselect","Bucket Sort","Suffix Array","Minimum Spanning Tree", "Counting Sort", "Shell",  "Line Sweep",  "Reservoir Sampling",   "Strongly Connected Component", "Eulerian Circuit",  "Radix Sort",  "Rejection Sampling",  "Biconnected Component"]], reader)

# print(dataset)


trainset, testset = train_test_split(dataset, test_size=0.2)

# Choose a similarity metric and build the model
sim_options = {'name': 'cosine', 'user_based': True}
model = KNNBasic(sim_options=sim_options)

# Train the model
model.fit(trainset)

# Make predictions
predictions = model.test(testset)

print(predictions)
# Evaluate the model
accuracy.rmse(predictions)
```

```python
from collections import defaultdict

# Create a dictionary to store recommended items for each user
recommendations = defaultdict(list)
print(recommendations)

# Iterate over the predictions
for prediction in predictions:
    uid = prediction.uid  # Get the user ID
    iid = prediction.iid  # Get the item ID (skill/topic)
    est = prediction.est  # Get the estimated rating
    
    # Store the recommended item for the user with its estimated rating
    recommendations[uid].append((iid, est))

# Sort the recommendations for each user based on the estimated rating
for uid, user_recs in recommendations.items():
    user_recs.sort(key=lambda x: x[1], reverse=True)

# Print recommendations for a specific user (change 'uid' to the desired user ID)
uid = 'iAmritSingh'
print(f"Recommendations for user {uid}:")
for item, est in recommendations[uid]:
    print(f"- {item}: {est}")
```
